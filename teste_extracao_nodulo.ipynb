# Extração do nódulo da imagem a partir da máscara

```python
import numpy as np
import pydicom
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img1 = pydicom.dcmread("D:\\ImagensDatasetFredrik\\00530_20990909_R_CC_1.dcm")
imagemArray1 = img1.pixel_array
masc1 = np.array(Image.open("D:\\ImagensDatasetFredrik\\mascaras\\00530_20990909_R_CC_1_mask.png"),dtype=int)
```

```python
x1 = np.array(imagemArray1)
x1
```

```python
nodulo = masc1 * x1
```

```python
plt.subplot(1,2,1)
plt.title(img1.PatientID + " - CranioCaudal")
plt.imshow(imagemArray1,cmap="gray")
plt.subplot(1,2,2)
plt.title(img1.PatientID + " - Localização do Nódulo")
plt.imshow(nodulo,cmap="gray");
plt.tight_layout()
```
