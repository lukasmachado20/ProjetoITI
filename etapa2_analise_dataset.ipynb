```python
import pandas as pd
import numpy as np
import os 
```

1. Data frame cronológico organizado de maneira temporal, de forma que saiba qual ano é cada imagem pro respectivo paciente.
    - Colunas: id_Paciente, nome_Imagem, ano_Imagem
    - Linhas: id_Paciente
2. Saber se, naquele respectivo ano, ele teve ou não teve cancer (1 ou 0).
3. Criar um csv com todos os pacientes que **tiveram câncer** e não tem máscara. Colocar colunas como **ID** e **ANO** que ele não teve máscara.

```python
def verifica_Mascaras(dataFrame,listaMascaras,caminhoArquivos):
    colMascaras = []
    arrayDf = np.array(dataFrame)
    for pac in arrayDf:
        # montando nome da mascara
        idPac = pac[1]
        lateralidade = pac[5]
        incidencia = pac[7]
        anoAcompanhamento = str(pac[2])
        caminhoMascara = idPac[:5] + "_20990909_" + lateralidade[0] + "_" + incidencia + "_" + anoAcompanhamento + "_mask.png"
        #print(caminhoMascara)
        # para agilizar, verificamos se este caminho da mascara está presente na lista
        if caminhoMascara not in listaMascaras:
            # não existe, então detalhamos no dataset 
            colMascaras.append('não existe')
        else:
            # existe, então detalhamos seu caminho e colocamos na lista
            colMascaras.append(caminhoMascara)
    return colMascaras
```

```python
dataFrame = pd.read_csv("anon_dataset_nonhidden_211125.csv") # lendo o dataset
```

```python
cancerOnlyDataframe = dataFrame[dataFrame['x_case']==1]
cancerOnlyDataframe
```

```python
cancerOnlyDataframe = cancerOnlyDataframe.loc[:,['anon_patientid','anon_filename','exam_year','x_case','x_cancer_laterality','imagelaterality','viewposition']]

# IMPORTANTE: Selecionei apenas as linhas onde a lateralidade do câncer é igual a imagem tomografada.
# x_cancer_laterality define em qual mama, esquerda ou direita, o câncer foi localizado.
# Por isso selecionar apenas as linhas que tem a mesma referência.
cancerOnlyDataframe = cancerOnlyDataframe.loc[ (cancerOnlyDataframe['imagelaterality']==cancerOnlyDataframe['x_cancer_laterality']) ,:]
cancerOnlyDataframe
```

- Percorrer o dataframe por paciente e colocar um número que identifique o acompanhamento do paciente a partir do ano das imagens. 

```python
arrayDf = np.array(cancerOnlyDataframe)
anoImg = arrayDf[0][2]
idPac = [0][0]
nroAcompanhamento = []
nro = 1
for pac in arrayDf:
    if pac[0] != idPac:
        # trocamos de paciente
        idPac = pac[0]
        anoImg = pac[2]
        nro=1
        nroAcompanhamento.append(nro) # recebe 1 no primeiro ano
    else:
        # estamos no mesmo paciente
        if anoImg != pac[2]: # se o ano for outro devemos somar +1 na contagem de anos
            anoImg = pac[2]  # atribui novo ano para as próximas verificações 
            nro+=1
            nroAcompanhamento.append(nro)
        else:
            # é o mesmo ano do verificado anteriormente
            nroAcompanhamento.append(nro) # coloca o mesmo número
            
# nova coluna dos acompanhamentos
cancerOnlyDataframe['anoAcompanhamento'] = nroAcompanhamento
```

```python
# conferindo
cancerOnlyDataframe = cancerOnlyDataframe.loc[:,['anon_patientid','anon_filename','anoAcompanhamento','exam_year','x_case','x_cancer_laterality','imagelaterality','viewposition']]
cancerOnlyDataframe.iloc[:20]
cancerOnlyDataframe['mascara'] = ''
cancerOnlyDataframe
```
