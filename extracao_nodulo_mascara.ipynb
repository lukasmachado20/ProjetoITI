```python
import numpy as np
import pydicom
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img1 = pydicom.dcmread("C:\\Users\\lukas\\Desktop\\Visão Computacional - UFCSPA\\Imagens DataSet\\00045_20990909_L_CC_1.dcm")
imagemArray1 = img1.pixel_array
masc1 = np.array(Image.open("C:\\Users\\lukas\\Desktop\\Visão Computacional - UFCSPA\\Imagens DataSet\\00045_20990909_L_CC_1_mask.png"),dtype=int)
```

```python
x1 = np.array(imagemArray1)
x1
```

```python
nodulo = masc1 * x1
```

```python
plt.subplot(1,2,1)
plt.title("Cranio Caudal")
plt.imshow(imagemArray1,cmap="gray")
plt.subplot(1,2,2)
plt.title("Localização do Nódulo")
plt.imshow(nodulo,cmap="gray")
```

```python
img2=pydicom.dcmread("C:\\Users\\lukas\\Desktop\\Visão Computacional - UFCSPA\\Imagens DataSet\\Mascaras\\00181_20990909_L_MLO_2.dcm")
img2Array=np.array(img2.pixel_array)
mascaraImg2= np.array(Image.open("C:\\Users\\lukas\\Desktop\\Visão Computacional - UFCSPA\\Imagens DataSet\\Mascaras\\00181_20990909_L_MLO_2_mask.png"),dtype=int)
noduloImg2= img2Array * mascaraImg2
```

```python
plt.subplot(1,2,1)
plt.title(img2.PatientID+" Médio Oblíqua -Lateral")
plt.imshow(img2Array,cmap="gray")
plt.subplot(1,2,2)
plt.title(img2.PatientID+" Nódulo")
plt.imshow(noduloImg2,cmap="gray")
```
